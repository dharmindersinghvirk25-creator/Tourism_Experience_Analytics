## 1.1.2 Data Description

### 1.1.2.1 Transaction Data:
**Purpose:** Contains information on user visits to various attractions, including ratings and visit details.  
**Columns:**  
TransactionId: Unique identifier for each transaction.  
UserId: Identifier for the user making the transaction.  
VisitYear: The year the visit occurred.  
VisitMonth: The month the visit occurred.  
VisitMode: The mode of visit (Business, Couples, Family, etc.).  
AttractionId: Unique identifier for the visited attraction.  
Rating: The user's rating for the attraction.  
**Usage:** This data will be used to understand user behavior, predict visit modes, and recommend attractions based on user preferences and ratings.  

### 1.1.2.2 User Data:
**Purpose:** Contains information about the users, such as their geographical location (continent, region, country, city).  
**Columns:**  
UserId: Unique identifier for each user.  
ContinentId: The continent to which the user belongs.  
RegionId: The region within the continent.  
CountryId: The country in which the user resides.  
CityId: The city of residence for the user.  
**Usage:** This data will help in analyzing the user demographics, which can be used to predict their behavior, visit modes, and attractions they might prefer.  
### 1.1.2.3 City Data:
**Purpose:** Contains information about different cities, used to link users and attractions to specific geographical locations.  
**Columns:**  
CityId: Unique identifier for each city.  
CityName: Name of the city.  
CountryId: The country associated with the city.  
**Usage:** This data helps to relate the CityId in the user and attraction data to a specific city name and country.  
### 1.1.2.4 Type Data:
**Purpose: **Contains details about type of tourist attractions.  
**Columns:**  
AttractionTypeId: The type of the attraction (e.g., Beach, Museum, Park).  
AttractionType: Type of the attraction.  
### 1.1.2.5 Visit Mode Data:  
**Purpose:** Contains information about the types of visit modes users may have, such as business trips, family vacations, or solo travel.    
**Columns:**    
VisitModeId: Unique identifier for the visit mode.    
VisitMode: Name or description of the visit mode (e.g., Business, Couples, Family).    
**Usage:** This data will be used to predict the user's likely visit mode based on their transaction history and demographic information.  
### 1.1.2.6 Continent Data:
**Purpose:** Contains information about the continents, helping to link users to their respective continents.  
**Columns:**  
ContinentId: Unique identifier for each continent.  
Continent: Name of the continent (e.g., Africa, Asia, Europe).  
**Usage:** Helps associate users with their continent, which may influence their travel behavior and preferences.  
### 1.1.2.7 Country Data:
**Purpose:** Contains information about countries, linking users and attractions to specific countries.  
**Columns:**  
CountryId: Unique identifier for each country.  
Country: Name of the country.  
RegionId: The region within the country.  
**Usage:** Helps in understanding user preferences and travel behavior in relation to different countries.  
### 1.1.2.8 Region Data:
**Purpose:** Contains information about regions within countries.  
**Columns:**  
RegionId: Unique identifier for each region.  
Region: Name or description of the region (e.g., North America, East Africa).  
ContinentId: Unique identifier for each continent.  
**Usage:** This data helps classify users and attractions based on regional preferences and trends.  
### 1.1.2.9 Item Data:  
**Purpose:** Contains information about regions within countries.  
**Columns:**  
AttractionId: Unique identifier for each attraction, linking it to other datasets.  
AttractionCityId: City where the attraction is located, connecting it to the City Data.  
AttractionTypeId: Category of the attraction (e.g., beach, historical site, park) for personalized recommendations.  
Attraction: Name of the attraction used to identify and suggest places.

## 2.3 Initial Missing Value Handling





```python
# Filling missing VisitMode with most frequent value
transaction['VisitMode'] = transaction['VisitMode'].fillna(transaction['VisitMode'].mode()[0])


# Filling missing Rating with average rating
transaction['Rating'] = transaction['Rating'].fillna(transaction['Rating'].mean())
```



## 2.4 Outlier Detection and Handling

```python
# Checking rating range
transaction['Rating'].describe()
```

```python
# Removing incorrect ratings (less than 1 or greater than 5)
transaction = transaction[(transaction['Rating'] >= 1) & (transaction['Rating'] <= 5)]
```

Outliers in Rating were removed by restricting values between 1 and 5 to ensure valid rating boundaries.

```python
# Converting VisitYear and VisitMonth to integer (if not already)
transaction['VisitYear'] = transaction['VisitYear'].astype(int)
transaction['VisitMonth'] = transaction['VisitMonth'].astype(int)
```

## 2.8 Dataset Integration (Merging All Tables)

Dataset Integration

All datasets are merged to create a single consolidated dataset containing:

*   User demographics
*   Visit details  
*   Attraction information
*   Location data


  


```python
# Merge transaction with user data
df = transaction.merge(user, on='UserId', how='left')

# Merge with city data
df = df.merge(city, on='CityId', how='left')

# Merge with item (type) data
df = df.merge(item, on='AttractionId', how='left')

# Merge with attraction type data
df = df.merge(type, on='AttractionTypeId', how='left')
```



## 2.10 Post-Merge Missing Value Handling

```python
# ==================================
# DATA PREPARATION – FINAL CLEANING
# ==================================


# Handle missing CityId
df['CityId'] = df['CityId'].fillna(df['CityId'].mode()[0])

# Handle missing CityName
df['CityName'] = df['CityName'].fillna("Unknown")

# Handle missing CountryId_y
df['CountryId_y'] = df['CountryId_y'].fillna(df['CountryId_y'].mode()[0])
```

### Explanation:

Missing values in location-related features were handled using mode imputation for categorical identifiers and “Unknown” labeling for missing city names to preserve dataset size.



## 2.11 Feature Engineering


```python
# 1 Previous average rating per attraction
df['AvgAttractionRating'] = df.groupby('AttractionId')['Rating'].transform('mean')

# 2 User average rating behavior
df['UserAvgRating'] = df.groupby('UserId')['Rating'].transform('mean')

# 3 Number of visits per user
df['UserVisitCount'] = df.groupby('UserId')['TransactionId'].transform('count')

# 4 Demographic aggregation - visits per city
df['CityVisitCount'] = df.groupby('CityId')['TransactionId'].transform('count')
```

### Explanation:

Feature engineering was performed to enhance model performance. Aggregated behavioral features such as average attraction rating, user average rating, visit frequency, and city-level visit count were created to capture user and attraction-level patterns.

## 4.3: Feature Engineering

Several derived features were created to enhance model performance:

*   Previous Average Rating per Attraction – Calculated mean rating for each attraction to capture historical performance trends.
*   User-Level Aggregated Features – Computed average rating given by each user to represent user satisfaction behavior.
*   Visit Frequency per User – Total number of visits made by each user to capture engagement level.
*   Demographic Aggregation – Integrated continent, region, country, and city data to represent geographical influence on travel behavior.

These engineered features improve predictive capability by capturing behavioral and contextual patterns beyond raw transactional data.


```python
# Feature Engineering - Creating Aggregated Features

# Previous Average Rating per Attraction
avg_rating_attraction = df.groupby('AttractionId')['Rating'].mean().reset_index()
avg_rating_attraction.columns = ['AttractionId', 'Avg_Attraction_Rating']
df = df.merge(avg_rating_attraction, on='AttractionId', how='left')

# User-Level Average Rating
avg_rating_user = df.groupby('UserId')['Rating'].mean().reset_index()
avg_rating_user.columns = ['UserId', 'Avg_User_Rating']
df = df.merge(avg_rating_user, on='UserId', how='left')

# Visit Frequency per User
visit_count = df.groupby('UserId')['TransactionId'].count().reset_index()
visit_count.columns = ['UserId', 'Visit_Frequency']
df = df.merge(visit_count, on='UserId', how='left')

print("Feature engineering completed successfully")
```

## 4.9 OBJECTIVE 3: RECOMMENDATION SYSTEM – Personalized Attraction Recommendation

### Aim:
To recommend relevant tourist attractions to users based on historical visit behavior, attraction characteristics, and user preferences.

### Machine Learning Technique Used:
User-Based Collaborative Filtering using Cosine Similarity

### Reason for Selection:
In this project, a user-based collaborative filtering approach is used to generate attraction recommendations based on similarity between users.

### Input Features:
UserId, AttractionTypeId, CityId, RegionId, CountryId, ContinentId, VisitHistory

### Output:
Top-N Recommended Attractions for each user

The recommendation system uses cosine similarity to identify similar users and suggest attractions based on shared preferences. This approach improves personalization and enhances overall user engagement.

## 4.9.1 Creating User–Item Matrix

```python
# Creating User-Attraction Matrix
user_item = df.pivot_table(
    index='UserId',
    columns='AttractionId',
    values='Rating'
).fillna(0)
```

```python
print("User-Item Matrix Shape:", user_item.shape)
```

## 4.9.2 Similarity Calculation

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity = cosine_similarity(user_item)
```

## 4.9.3 Recommendation Function

```python
def recommend_attractions(user_id, top_n=5):
    user_idx = user_item.index.get_loc(user_id)
    similarity_scores = list(enumerate(user_similarity[user_idx]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    similar_users = [i[0] for i in similarity_scores[1:top_n+1]]

    recommended = user_item.iloc[similar_users].mean().sort_values(ascending=False)
    return recommended.head(top_n)
```

## 4.9.4 Example Output

```python
recommend_attractions(user_item.index[0])
```

## 4.10 Recommendation Evaluation (RMSE)

```python
from sklearn.metrics import mean_squared_error
import numpy as np

actual = user_item.values.flatten()
predicted = user_similarity.dot(user_item).flatten()

rmse = np.sqrt(mean_squared_error(actual, predicted))

print("Recommendation RMSE:", rmse)
```
